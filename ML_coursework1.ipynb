### custom helper function

```python
def np_round(x, y):
    return round(x, y)


def cal_threshold(func, test_y, pred_y, thresholds, print_mse_values = False):
    mse_values = []
    for threshold in thresholds:
        # Calculate MSE for the current threshold
        mse_values.append(func(test_y, pred_y > threshold))

    if print_mse_values:
      print(f"{mse_values =}")

    # Find the index of the minimum MSE
    min_mse_index = np.argmin(mse_values)
    # Update the threshold with minimum MSE
    min_mse_threshold = thresholds[min_mse_index]

    return min_mse_threshold, min_mse_index


def threshold_opt(func, test_y, pred_y, start=0.0, end=0.1, n_decimal=4):
    min_mse = []
    thresholds = np.linspace(start, end, 10)

    for _ in range(n_decimal):
        mse, idx = cal_threshold(func, test_y, pred_y, thresholds)

        if len(min_mse) > 0 and min_mse[-1] < mse:
            break

        print(f"Iteration: {_}, Threshold: {mse}, MSE: {mse}")
        min_mse.append(mse)

        # Update thresholds for the next iteration
        thresholds = np.linspace(mse, mse + (end - start) / 10, 10)

    return min_mse

```

## 2.8 Data Pipeline

```python
from platform import processor
from sklearn.base import BaseEstimator, TransformerMixin

class LabelEncoderTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.label_encoder = LabelEncoder()

    def fit(self, X, y=None):
        self.label_encoder.fit(X)
        return self

    def transform(self, X):
        if isinstance(X, pd.Series):
            return pd.DataFrame(self.label_encoder.transform(X), columns=[X.name])
        else:
            return self.label_encoder.transform(X).reshape(-1, 1)




def data_pipeline_features(X):
    X1 = X.copy()

    # Convert "Distance" column to numeric type
    X1["Distance"] = pd.to_numeric(X1["Distance"], errors='coerce')

    # Define preprocessing steps for different columns
    # numeric_features = ['Distance']
    numeric_features = infrared_thermography_temperature.variables.name[(infrared_thermography_temperature.variables.type == "Continuous")][2:]

    categorical_features = ['Gender', 'Age', 'Ethnicity']
    gender_categories = [['Female','Male']]

    # Define transformers for preprocessing
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),  # Impute missing values with median
        ('standardScalar', StandardScaler())
    ])

    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values with most frequent value
        ('encoder', OneHotEncoder(sparse=True, drop="if_binary", categories=gender_categories)),  # One-hot encode categorical variables
    ])

    ordinal_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values with most frequent value
        ('encoder', OrdinalEncoder()),  # Ordinal encode categorical variables
    ])

    label_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values with most frequent value
        ('encoder', LabelEncoderTransformer()),  # Label encode categorical variables
    ])

    # Combine transformers based on data type
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat_onehot', categorical_transformer, ['Gender']),
            ('cat_ordinal', ordinal_transformer, ['Age']),
            ('cat_label', label_transformer, ['Ethnicity']),
            ('num', numeric_transformer, numeric_features),
        ])

    # Define the pipeline
    pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

    # Fit and transform the data
    X_processed = pipeline.fit_transform(X1)

    # columns for the Dataframe
    columns = categorical_features + list(numeric_features)

    return  pd.DataFrame(X_processed, columns = columns)




def data_pipeline_target(y):
    # copy the target data frame
    y1 = y.copy()

    # Define Transformer for preprocessing
    target_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy="median"))
    ])

    # Fit and Transform the data
    y_processed = target_transformer.fit_transform(y1)

    # Add the fever column
    fever = (y1["aveOralM"] >= 37.5)
    y_processed = pd.DataFrame(y_processed, columns=y.columns)
    y_processed["fever"] = (np.zeros(y.shape[0]) + fever).astype("uint8")

    return y_processed




def data_pipeline(X=None, y=None):
    X_processed = None
    y_processed = None

    if X is not None and not X.empty:
        X_processed = data_pipeline_features(X)

    if y is not None and not y.empty:
        y_processed = data_pipeline_target(y)

    return X_processed, y_processed
```

## 3.1 Finding the common baseline

### a) Common sense base line for the regression



**Regression Task (Predicting Oral Temperature):**
